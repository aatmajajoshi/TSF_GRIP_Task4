## Import the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcdefaults()
plt.rcParams.update({'axes.facecolor':'white'})
```

## Load the Dataset

```python
df = pd.read_csv('globalterrorismdb_0718dist.csv', encoding='latin1', engine='python')
df.head()
```

## Data Preprocessing

```python
## Columns having null values
df.isnull().sum()
```

```python
## Renaming the columns
df.rename(columns = {'iyear':'Year', 'imonth':'Month', 'country_txt':'Country', 'region_txt':'Region', 'provstate':'State', 'attacktype1_txt':'Attack Type', 'targtype1_txt':'Target Type', 'targsubtype1_txt':'Target Subtype','target1':'Target', 'gname':'Group Name', 'weaptype1_txt':'Weapon Type', 'weapsubtype1_txt':'Weapon Subtype','nkill':'Killed', 'nwound':'Wounded', 'propcomment':'Property Damage'}, inplace = True)
```

```python
df= df[['Year','Month','Country','Region','State','city','location','success','Attack Type','Target Type','Target Subtype','Target','Group Name','motive','Weapon Type','Weapon Subtype','Killed','Wounded','Property Damage']]
df.head()
```

```python
# Now we can change the shape
df.shape
```

```python
ctry = df['Country'].value_counts()[:50].plot.bar(figsize = (22, 10), color = 'Darkcyan')
plt.xlabel('Country', fontsize = 16)
plt.xticks(fontsize=16)
plt.title('Top 50 terrorism-prone countries', fontsize = 25)
plt.show()
```

```python
region = df['Region'].value_counts()
region.plot.bar(figsize = (22,8), color = 'Maroon')
plt.xlabel('Region', fontsize = 16)
plt.xticks(fontsize = 16)
plt.title('Terrorism Prone Regions', fontsize = 25)
plt.show()
```

```python
target_type = df['Target Type'].value_counts()
target_type.plot.bar(figsize = (22,8), color = 'Orange')
plt.xlabel('Target Type',fontsize = 16)
plt.xticks(fontsize = 16)
plt.title('Targetted Entities', fontsize = 25)
plt.show()
```

```python
# Filter out 'Unknown' city beforehand if possible
filtered_df = df[df['city'] != 'Unknown']

# Count the occurrences of each city and select the top 20
top_cities = filtered_df['city'].value_counts().head(20)

# Plot the top cities
top_cities.plot.barh(figsize=(22, 8), color='blue')
plt.ylabel('City', fontsize=16)
plt.yticks(fontsize=16)
plt.title('Top 20 Cities Affected by Attacks', fontsize=25)
plt.gca().invert_yaxis()
plt.show()
```

```python
# Replace empty values with 'Unknown' if needed
df['location'].fillna('Unknown', inplace=True)

# Filter out 'Unknown' locations beforehand if possible
filtered_df = df[df['location'] != 'Unknown']

# Count the occurrences of each location and select the top 50
top_locations = filtered_df['location'].value_counts().head(50)

# Plot the top locations
top_locations.plot.barh(figsize=(15, 20), color='brown')
plt.ylabel('Location', fontsize=16)
plt.yticks(fontsize=16)
plt.title('Top 20 Locations Affected by Attacks', fontsize=25)
plt.gca().invert_yaxis()
plt.show()
```

```python
# Replace NaN and 'Unknown' values with a consistent label, such as 'Not Specified'
df['State'].fillna('Not Specified', inplace=True)
df['State'].replace('Unknown', 'Not Specified', inplace=True)

# Filter out 'Not Specified' states beforehand if possible
filtered_df = df[df['State'] != 'Not Specified']

# Count the occurrences of each state and select the top 50
top_states = filtered_df['State'].value_counts().head(50)

# Plot the top states
top_states.plot.barh(figsize=(15, 20), color='darkgreen')  
plt.ylabel('State', fontsize=16)  
plt.yticks(fontsize=12)  
plt.title('Top 50 States Affected by Attacks', fontsize=20)  
plt.gca().invert_yaxis() 
plt.show()
```

```python
plt.figure(figsize = (20,8))
sns.set_theme(style = "ticks")
sns.histplot(df, x = "Year", hue = "Region", multiple = "stack", palette = "light:m_r", edgecolor = ".3", linewidth = .5, log_scale = False)
plt.xlabel('Year', fontsize = 16)
plt.xticks(fontsize = 16)
plt.title('Region-Wise Terrorism over the years', fontsize = 25)
plt.show()
```

```python
df.groupby('Attack Type')[['Killed', 'Wounded']].count()
```

```python
attack = df.groupby('Attack Type')[['Killed', 'Wounded']].count()
attack.plot.bar(color = ['red', 'yellow'])
plt.show()
```

```python
GName = df['Group Name'].value_counts().drop('Unknown')[:20]
GName.plot.barh(figsize = (22, 15))
plt.ylabel('Group Name', fontsize = 16)
plt.yticks(fontsize = 16)
plt.title('Top 20 Terrorist Groups', fontsize = 25)
plt.gca().invert_yaxis()
plt.show()
```
